```python
# Reducing / Replacing the Unknowns 
# Identify unknowns and known countries
unknown_mask = df['Country/Region'].str.lower() =='unknown'
known_countries = df.loc[~unknown_mask, 'Country/Region'].unique()

# Get total unknown count
total_unknown = unknown_mask.sum()
print(f"Total unknowns: {total_unknown}")
```

```python
# Calculate how many to redistribute (95%)
redistribute_count = int(total_unknown * 0.95)

# Randomly choose which unknown rows to redistribute
redistribute_indices = np.random.choice(df[unknown_mask].index, size=redistribute_count, replace=False)
```

```python
# Randomly assign them to known countries
df.loc[redistribute_indices, 'Country/Region'] = np.random.choice(known_countries, size=redistribute_count, replace=True)
```

```python
# The rest (20%) remain 'Unknown'
remaining_unknown = total_unknown - redistribute_count
print(f"Redistributed: {redistribute_count} | Remaining Unknown: {remaining_unknown}")
```
